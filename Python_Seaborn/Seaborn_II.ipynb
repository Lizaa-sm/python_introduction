## **Seaborn - Python for Data Visualization**

Here we go again.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data= tips, estimator= np.std)
sns.countplot(x = 'sex', data= tips, )
```

```python
sns.boxplot(x= 'day', y= 'total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data= tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='Black')
```

```python
# You can also choose the categorical plot type with kind:
# sns.catplot(x='day', y='total_bill', data=tips, kind='box')     # boxplot
# sns.catplot(x='day', y='total_bill', data=tips, kind='violin')  # violinplot
sns.catplot(x= 'day', y='total_bill', data=tips, kind='bar', color= 'pink')
```

* Finally, you can use ``catplot`` to call the graph and define the style with ``kind``, which can be any one already shown directly.

## **Matrix Plots**

* Continuing...

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
# loads the built-in dataset "flights", which contains airline passenger counts by month and year.
flights = sns.load_dataset('flights') 
tips.head()
```

```python
flights.head()
```

* Creating ``heatmaps`` requires data in a matrix format (rows × columns). This matrix can usually come from two main sources: a correlation matrix, which shows numerical relationships between variables, or a pivot table, which organizes data into categorical ``rows`` and ``columns``. The ``sns.heatmap()`` function relies on this grid-like structure to generate the visualization, where colors represent the data values.

```python
ex = tips.corr(numeric_only=True)
ex
```

```python
# you take notes annotation
sns.heatmap(ex, annot=True)
```
