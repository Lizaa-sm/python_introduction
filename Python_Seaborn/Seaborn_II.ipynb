## **Seaborn - Python for Data Visualization**

Here we go again.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data= tips, estimator= np.std)
sns.countplot(x = 'sex', data= tips, )
```

```python
sns.boxplot(x= 'day', y= 'total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data= tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='Black')
```

```python
# You can also choose the categorical plot type with kind:
# sns.catplot(x='day', y='total_bill', data=tips, kind='box')     # boxplot
# sns.catplot(x='day', y='total_bill', data=tips, kind='violin')  # violinplot
sns.catplot(x= 'day', y='total_bill', data=tips, kind='bar', color= 'pink')
```

* Finally, you can use ``catplot`` to call the graph and define the style with ``kind``, which can be any one already shown directly.
