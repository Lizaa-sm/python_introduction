```python
def ch_s(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
ch_s('Hi morning CHIEF here')
```

```python
# The people have the word Chief in their job title:

import re

# Mask com regex que procura a palavra whole-word 'chief' (case-insensitive)
mask = salaries['JobTitle'].str.contains(r'\bchief\b', case=False, regex=True, na=False)

# Quantidade
count = mask.sum()
print("Count:", count)

# Linhas que correspondem
result_df = salaries[mask]
result_df.head()

# Se quiser ver quais tÃ­tulos aparecem e quantas vezes:
print(result_df['JobTitle'].value_counts())
result_df[:15]
```
