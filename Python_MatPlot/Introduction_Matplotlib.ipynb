## **MatPlotlib - Introduction**

* Matplotlib is a Python library used for creating plots and data visualizations. It allows you to generate anything from simple line charts to more complex visualizations, such as histograms, bar charts, scatter plots, and more. It is highly flexible and often used alongside libraries like NumPy and Pandas for data analysis.

```python
import matplotlib.pyplot as plt

%matplotlib inline 

# This is a Jupyter Notebook magic command. It makes plots generated by matplotlib show up directly 
# inside the notebook, right below the code cell. Without it, plots might open in a separate window.
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y)
# plt.show()
# If you want change color: 
plt.plot(x, y, 'r-')
# If you want to add a label, just do:
plt.xlabel('x - label')
plt.ylabel('y - label')
# If you want to add a title, do this:
plt.title('Title Example')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r-')
```

```python
# object-oriented programming (OOP)
fig_1 = plt.figure()

axes = fig_1.add_axes([0.5,0.5,1.0,1.0])

axes.plot(x, y, 'r-')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title Yes')
```

```python
fig_2 = plt.figure()

# Now it is possible to improve the frame by axes
axes1 = fig_2.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.set_title('Plot 1')

axes2 = fig_2.add_axes([0.5, 0.15, 0.4, 0.5])
axes2.set_title('Plot 2')
```
